```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eulers(f,h,a,b,y0):
    N = int((b-a)/h)
    mesh = [a+(i*h) for i in range(N+1)]
    
    y = [y0]
    for t in mesh: 
        if t==a: continue
        y_i = y0 + (h*f(t,y0))
        y.append(y_i)
        y0 = y_i
    
    return mesh, y
```

```python
def f(t,y):
    return (-5*y + 5*(t**2) + 2*t)

def y(t):
    return (t**2 + (1/3)*np.exp(-5*t)) 

x = np.linspace(0,1,100)
y = y(x)

x1,y_01 = eulers(f,0.1,0,1,1/3)
x2,y_005 = eulers(f,0.05,0,1,1/3)
x3,y_0025 = eulers(f,0.025,0,1,1/3)

plt.figure(dpi=150)
plt.plot(x,y, label='true soln', lw=2, c='black', ls='--')
plt.plot(x1,y_01, label='h=0.1')
plt.plot(x2,y_005, label='h=0.05')
plt.plot(x3,y_0025, label='h=0.025')
plt.legend()
plt.show()
```

```python
def aitken_est(y1,y2,y3):
    return np.log((y1-y2)/(y2-y3)) / np.log(2)
```

```python
aitken_est(y_01[-1],y_005[-1],y_0025[-1])
```
